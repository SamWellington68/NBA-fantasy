```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np 
data =pd.read_csv("inst126basketballstats1.csv",)
adp=pd.read_csv("adpfix.csv")
data["Fantasy Total"]=data['PTS']+(data['TRB']*1.2)+(data['AST']*1.5)+ (data['BLK']*2)+(data['STL']*2)#algorithm to find "best" players
key_cols = ['Player']
data = data.merge(adp[key_cols + ['ADP']], how='outer', on=key_cols) #we asked a question on stack exchange and got this as the answer, we had been stuck for days and had attended multiple office hours
data.sort_values('ADP',axis=0,ascending=True,inplace=True,na_position='last')
position=input("what position do you need? ") 
position = position.casefold() 
if position == "pg":
    point = data.loc[(data['Pos']=='PG')]
    print("Here are the best fantasy point guards ")
    data1=(point.sort_values('Fantasy Total', ascending=False)) # sorts data by best players based on algorithm above
    graphs()
if position == "sg":
    point = data.loc[(data['Pos']=='SG')]
    print("Here are the best fantasy shooting guards")
    data1=(point.sort_values('Fantasy Total', ascending=False))
    graphs()
if position == "sf":
    point = data.loc[(data['Pos']=='SF')]
    print("Here are the best fantasy small forwards")
    data1=(point.sort_values('Fantasy Total', ascending=False))
    graphs()
if position == "pf":
    point = data.loc[(data['Pos']=='PF')]
    print("Here are the best fantasy power fowards")
    data1=(point.sort_values('Fantasy Total', ascending=False))
    graphs()
if position == "c":
    point = data.loc[(data['Pos']=='C')]
    print("Here are the best fantasy centers")
    data1=(point.sort_values('Fantasy Total', ascending=False))
    graphs()
```
